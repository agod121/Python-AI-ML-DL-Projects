```python
import pandas as pd
```

```python
# Setting the start and end dates for pulling the stock prices
start_date = '2015-01-01'
end_date = '2022-07-25'
```

```python
# Assigning google stock ticker to a variable 
ticker = 'TSLA'
```

```python
# Importing yahoo as finance 
import yfinance as yf 
```

```python
# Getting the data for the ticker between the start and end dates
data = yf.download(ticker, start = start_date, end = end_date)
```

```python
data
```

```python
# Adding the index/row number to the dataset
data.reset_index(inplace=True)
data
```

```python
# Selecting the Date and the Adjusted Closing Price columns from the main dataset
df = data[["Date", "Adj Close"]]
df
```

```python
# Creating a a datestamp column to store the date values
df['ds'] = df['Date']
df
```

```python
# Creating a a datestamp column to store the date values
df['y'] = df['Adj Close']
df
```

```python
# Importing the prophet function the prophet package
from prophet import Prophet
```

```python
#Creating a prophit instance. If it is a pure python package you do not need to create an instance of the package
m = Prophet()
```

```python
# Fitting the forecasting model to our data 
m.fit(df)
```

```python
# Creating a dataframe for predicting the values for 20 periods
future = m.make_future_dataframe(periods=20)
```

```python
# Storing prophet's predictions in a variable called forecast
forecast = m.predict(future)
```

```python
forecast
```

```python
# Building a predictions dataset
predictions = forecast[["ds", "yhat", "yhat_lower", "yhat_upper"]]
predictions
```

```python
# Downloading the predictions to a local file
predictions.to_csv("Google Stock Predictions.csv")
```
