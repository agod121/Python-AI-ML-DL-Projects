```python
# Loading the basic packages
import numpy as np
```

```python
# Importing the Resnet50 from deep learning package - Keras
from keras.applications.resnet import ResNet50
```

```python
# Importing the image processing package
from keras.preprocessing import image
```

```python
# Setting ResNet50 as our classifier
classifier = ResNet50()
```

```python
# Loading the image for object recognition
new_image = image.load_img('/content/Ape.jpg', target_size = (224,224))
new_image
```

```python
# Transforming the image to an array
transformed_image = image.img_to_array(new_image)
transformed_image
```

```python
# Expanding the dimensions with a numpy package
transformed_image = np.expand_dims(transformed_image, axis = 0)
transformed_image
```

```python
# Scaling the image
from keras.applications.resnet import preprocess_input
transformed_image = preprocess_input(transformed_image)
transformed_image
```

```python
# Predicting the image using Resnet50
y_pred = classifier.predict(transformed_image)
y_pred
```

```python
# Decoding the predictions
from keras.applications.resnet import decode_predictions
decode_predictions(y_pred, top=20)
label = decode_predictions(y_pred)
label
```
